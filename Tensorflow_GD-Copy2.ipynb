### Short introduction to Gradient descent.

<br />
In this coding example we explain how gradient descent would work for a single layer preceptron with linear output.

```python
import numpy as np
import matplotlib.pyplot as plt
import pandas as pd
import math
import os
```

```python
import tensorflow.compat.v1 as tf1
#For Compatibility with older code.
tf1.disable_eager_execution()
```

```python
data = pd.read_csv('https://vincentarelbundock.github.io/Rdatasets/csv/carData/Ginzberg.csv', index_col=0)
```

```python
dx = data.values
```

```python
X = dx[:,:-1]#np.array([0.5,2.3,2.9])
Y = dx[:,-1]#np.array([1.4,1.9,3.2])
```

```python
# This time we randomize both variables needed, also we enclose them inside an array
params = tf1.Variable([list(np.random.uniform(size = (6)))],dtype=tf1.float32)
```

```python
Xmat = np.hstack([np.ones(len(X)).reshape(-1,1),X])
Xmat = Xmat.astype(np.float32)
```

```python
y_hat = tf1.matmul(params,Xmat.T)
error = tf1.reduce_mean((Y - y_hat)**2)
```

```python
optimizer = tf1.train.GradientDescentOptimizer(learning_rate = 0.1)
train = optimizer.minimize(error)
init = tf1.global_variables_initializer()
```

```python
with tf1.Session() as sess:
    sess.run(init)
    training_steps = 15000
    for i in range(training_steps):
        sess.run(train)
        p =  sess.run(params)
        print (sess.run(error), p)
        SumSq = sess.run(error)
```
